```python
import pandas as pd
import xml.etree.ElementTree as ET
import pickle
import os
from os import listdir, getcwd
from os.path import join
```

```python
df = pd.read_csv('avgerror.txt',names = ['data'])
```

```python
A = df['data'].str[:].str.split(' ', expand=True).astype(float)
```

```python
index = A[0].values.astype(int)
loss = A[1].values
avtloss = A[2].values
```

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
loss_downsample = np.arange(802)
avg_downsample = np.arange(802)
index_downsample = np.arange(802)
for i in range(80200):
    loss_downsample[int(i / 100)] = loss[i]
    avg_downsample[int(i / 100)] = avtloss[i]
    index_downsample[int(i / 100)] = index[i]
```

```python
plt.figure(figsize=(10,5))
plt.plot(index_downsample,loss_downsample,'c')
plt.xlabel('Iteration')
plt.ylabel('Loss')
```

```python
plt.figure(figsize=(10,5))
plt.plot(index_downsample,avg_downsample,'m')
plt.xlabel('Iteration')
plt.ylabel('AverageLoss')
```
